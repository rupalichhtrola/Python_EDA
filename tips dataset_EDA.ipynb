# Data Visualization


```python
plt.figure(figsize=(10,6))
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='smoker',data=df,palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day',data=df,palette='RdBu_r')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='time',data=df,palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day',data=df,hue='time',palette='RdBu_r')
```

```python
x=df['total_bill']
y=df['tip']
plt.scatter(x,y)
plt.xlabel('Total Bill',fontsize=15)
plt.ylabel("Tip",fontsize=15)
```

```python
plt.hist(df['tip'])
```

```python
df.day.value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='total_bill',data=df,palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='time',y='tip',data=df)
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='time',y='tip',hue='day',data=df)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['total_bill'])
```

```python
corr_matrix=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap="YlGnBu")
```
